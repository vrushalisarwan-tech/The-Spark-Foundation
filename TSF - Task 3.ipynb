```python
df=df.drop(['Country','Postal Code'],axis=1)
```

```python
df_numeric = df.select_dtypes(include='number')
df_numeric.plot(kind = 'kde',subplots = True, layout = (2,2), sharex = False, figsize = (10,10))
plt.grid(True)
plt.show()
```

```python
sns.pairplot(df)
```

```python
sns.distplot(df["Sales"])
```

```python
x = df['Segment']
y = df['Sales']
plt.bar(x,y, color = 'green')
plt.title('Segments vs Sales')
plt.xlabel('Segments')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize =(5,5))
df['Segment'].value_counts().plot(kind='pie',title='Various Segments',legend=True,autopct='%1.1f%%')
```

```python

plt.figure(figsize =(7,7))
df.groupby('Region')['Profit'].sum().plot(kind = 'pie',autopct='%1.1f%%',explode=(0, 0, 0,0.1))
plt.show()
```

```python
df.groupby('Region')['Sales','Profit'].sum().plot.bar()
```

```python

plt.figure(figsize =(20,10))
df.groupby('Sub-Category')['Sales','Profit'].sum().sort_values(by = ['Profit'] , ascending = True).plot.bar()
plt.ylabel('Sales')
```

```python
df[df['Category']=='Furniture'].groupby('Sub-Category')['Sales','Profit'].sum().plot.bar()
```

```python
df.groupby('State')['Quantity'].sum().sort_values(ascending = False).plot.bar(figsize = (15, 5))
plt.title('Total State-Wise Deals')
plt.ylabel('Total Number of Quantities')
plt.xlabel('American States')
plt.show()
```

```python
plt.figure(figsize =(20,10))
df.groupby('State')['Profit'].sum().sort_values(ascending = False).plot(kind = 'bar')
plt.title('States vs Profit' , fontsize = 20)
plt.grid(True)
plt.show()
```

```python
# Sales vs Profit
sns.scatterplot(x= 'Sales', y ='Profit', data =df)
plt.grid(True)
plt.show()
```

```python
sales_discount = pd.pivot_table(data = df, index = 'Region', columns = 'Discount', values = 'Sales', aggfunc = 'sum')
sns.heatmap(sales_discount, cmap = 'Reds')
plt.show()
```

```python
sales_sub_discount = pd.pivot_table(data = df, index = 'Sub-Category', columns = 'Discount', values = 'Sales', aggfunc = 'sum')
sns.heatmap(sales_sub_discount, cmap = 'Greens')
plt.show()
```

```python
df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending =False).plot.bar(color='g')
plt.ylabel('Profit')
```

```python
plt.figure(figsize = (8, 5))
# profit/loss by Discount level
sns.lineplot('Discount', 'Profit', data = df, color = 'green', label = 'Discount Level')
plt.ylabel('Profit')
plt.title('Profit/Loss by comparing Discount Level', fontsize = 20)
plt.show()
```

Conclusions : 

- The discount level should be lowered to 10% to increase the profit.
- The sales should be focused equally on all states in order to reach consumer satisfactory levels.
- The worst losses occured when discount approximates 50%. This may indicate end-of-season sales, and sales for older designs or defect in their products.
- Superstore is most profitable when the discount level reaches lower than 10%. This allows less occurences of lowered prices after decreasing the discount level, and thus more profits.
- We need to reduce the sales of furniture as it is giving less profit for more no. of sales.
